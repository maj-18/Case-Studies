# 3. Graph of male vs female buyers participated in the sales

```python
#Count of gender
gnd=datas['Buyer Gender'].value_counts()
gnd
```

```python
plt.figure(figsize=(10,8))
plt.title('Male vs Female buyers participated in the sales',fontsize='15')
lab=sns.countplot(x='Buyer Gender',data=datas)
for p in lab.patches:
        lab.text(p.get_x()+.15,p.get_height()-300,str(p.get_height()),fontsize=18,ha='center',color='black',fontweight='bold')
```
